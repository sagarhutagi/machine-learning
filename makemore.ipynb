```python
# Getting the name from the .csv file 
n_list = [first.split(',')[0].split(' ')[0] for first in open('male_names.csv', 'r').read().splitlines()]
name_list = n_list[1:]
```

```python
sample = name_list[100:105]
sample
```

```python
b = {}
for w in name_list:
    chs = ['<S>']+list(w)+['<E>']
    for i in range(len(chs)-1):
        biagram = (chs[i]+chs[i+1])
        b[biagram] = b.get(biagram, 0) + 1



# print(sorted(b.items(), key= lambda kv : kv[1], reverse=True))
```

```python
import torch
```

```python
a = torch.zeros((79,79), dtype=torch.int32)
a
```

```python
chars = (sorted(set(''.join(name_list))))
stoi = {s:i+1 for i,s in enumerate(chars)}
stoi['.'] = 0
stoi
```

```python
for w in name_list:
    chs = ['.']+list(w)+['.']
    for i in range(len(chs)-1):
        ix1 = stoi[chs[i]]
        ix2 = stoi[chs[i+1]]
        a[ix1,ix2] += 1
```
