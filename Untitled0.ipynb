```python
import requests
name_git_link = "https://gist.githubusercontent.com/mbejda/7f86ca901fe41bc14a63/raw/38adb475c14a3f44df9999c1541f3a72f472b30d/Indian-Male-Names.csv"
text = requests.get(name_git_link).text
all_names = [
    line.split(",")[0].split()[0]
    for line in text.splitlines()[1:]
    if line.strip() and line.split(",")[0].strip()
]
```

```python
import re
names = [n for n in all_names if re.fullmatch(r"[A-Za-z]+", n)]
```

```python
names[100:110]
```

```python
len(names)
```

```python
import torch
N = torch.zeros((27,27), dtype=torch.int32)
```

```python
chars = sorted(list(set(''.join(names))))
stoi = {s:i+1 for i,s in enumerate(chars)}
itos = {i+1:s for i,s in enumerate(chars)}
stoi["."] = 0
```

```python
for w in names:
  chs = ["."] + list(w) + ["."]
  for ch1, ch2 in zip(chs, chs[1:]):
    N[stoi[ch1], stoi[ch2]] += 1
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.imshow(N)
```

```python
# g = torch.Generator().manual_seed(2147483647)

for i in range(20):
  out = []
  ix = 0
  while True:
    p = N[ix].float()
    p = p/p.sum()
    ix = torch.multinomial(p, num_samples=1, replacement=True).item()
    if ix == 0:
      break
    out.append(itos[ix])
  print(''.join(out))
```
